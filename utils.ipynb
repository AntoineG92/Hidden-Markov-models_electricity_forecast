## Add daytypes

```python
import pandas as pd
from workalendar.europe import France
import calendar
from datetime import date, timedelta
```

```python
cal = France()
```

```python
df=pd.read_csv("cdc_conso.csv",sep=";")
print(df.shape)
df.head()
```

```python
#check types of dataframes columns
df.dtypes
```

```python
#transform the targeted column values into datetime
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
#Create new columns Daytypes :
#0 : monday
#0 : tuesday / wednesday / thursday
#0 : friday
#0 : saturday
#0 : sunday

df['daytype0'] = df['Date'].dt.dayofweek == 1
df['daytype1'] = df['Date'].dt.dayofweek.isin([2,3,4])
df['daytype2'] = df['Date'].dt.dayofweek == 5
df['daytype3'] = df['Date'].dt.dayofweek == 6
df['daytype4'] = df['Date'].dt.dayofweek == 7

#We look at the results
df
```

```python
#bank holidays
def bank_holiday(timestamp):
    date = timestamp.to_pydatetime()
    return cal.is_holiday(date)


df['daytype6'] = df['Date'].apply(bank_holiday)
df[df['daytype6']==True]
```

```python
#before bank holidays
def before_bank_holiday(timestamp):
    is_before_bank_holiday = False
    tomorrow_date = timestamp.to_pydatetime()+ timedelta(days=1)
    tomorrow_timestamp = pd.Timestamp(tomorrow_date)
    if bank_holiday(tomorrow_timestamp) and (not bank_holiday(timestamp)):
        is_before_bank_holiday=True
    return is_before_bank_holiday

df['daytype5'] = df['Date'].apply(before_bank_holiday)
df[df['daytype5']==True]
```

```python
#after bank holidays
def after_bank_holiday(timestamp):
    is_after_bank_holiday = False
    yesterday_date = timestamp.to_pydatetime() - timedelta(days=1)
    yesterday_timestamp = pd.Timestamp(yesterday_date)
    if bank_holiday(yesterday_timestamp) and (not bank_holiday(timestamp)):
        is_after_bank_holiday=True
    return is_after_bank_holiday

df['daytype7'] = df['Date'].apply(after_bank_holiday)
df[df['daytype7']==True]
```

```python
#between bank holidays and weekend
def between_bank_holiday_and_weekend(timestamp):
    is_between_bank_holiday_and_weekend = False
    today_date = timestamp.to_pydatetime()
    if (today_date.weekday()==5 and after_bank_holiday(timestamp)) or today_date.weekday()==1 and (before_bank_holiday(timestamp)):
            is_between_bank_holiday_and_weekend=True
    return is_between_bank_holiday_and_weekend

df['daytype8'] = df['Date'].apply(between_bank_holiday_and_weekend)
df[df['daytype8']==True]
```
